image.png

image.png

1. we dont look for exact relation but just relation
2. focus on worst case scenarios so use nested not for because nested taek more time
3. this worst case is called big Oh O()

```python
#factorial function
def fact_iter(n):
  answer = 1      #first operation
  while n > 1:    #second operation (comparision)
    answer *= n   #two operation
    n -= 1        #two operations
  return answer

fact_iter(4)
#now how to calculate the time complexity of this program?
# 1 + 5n
```

How
1. Count operations
    1 for answer
    5n for loop
2. So it will be 1 + 5n
3. now for calculating O()
4. remove additive constant 1
5. then  remove multiplicative constant 5
6. answer is O(n)

```python
#problem no 3 : integers into strings
def intostr(i):
  degits = '0123456789'
  if i==0:
    return 0
  result=''
  while i > 0:
    result = degits[i%10] + result  #1%10->1, 2%10->2 ....10%10->0
    i = i//10    #1->0, 2->0, 3->0, ...10->1
  return result


intostr(123)
#visualization on pythontutor.com
```

```python
#Problem no 06 : Binary Search
L=[1,2,3,4,5]

for i in range(0,len(L)): #five times
  for j in range(i+1,len(L)):  #four times
    print('({},{})'.format(L[i],L[j])) #O(n)*O(n)

#O(n^2)
```

```python
#Problem no 07
A=[1,2,3,4]
B=[2,3,4,5,6]

for i in A: #O(a)
  for j in B: #O(b)
    if i<j: #O(1)
      print(('{},{}').format(i,j)) #O(n)*O(n)
#O(ab) <- better but is also correct O(n^2)
```

```python
#Problem no 07
A=[1,2,3,4]
B=[2,3,4,5,6]

for i in A:
  for j in B:
    for k in range(100000):
      print(('{},{}').format(i,j))
# O(n^2) because in the third loop we have a constant number
```

```python
#Problem no 08 : it will reverse the list you entered
L=[1,2,3,4]

for i in range(0,len(L)//2):
  other=len(L)-i-1
  temp=L[i]
  L[i]=L[other]
  L[other]=temp

print(L)
#Time complexity is O(n)
```

```python
#Now for Recurssion
#Problem no 09 :
def factorial(n):
  if n==1:
    return 1
  else:
    return n*factorial(n-1)

print(factorial(5))
#its time coplexity is O(n) linear
```

```python
#Problem no 10 : Fibonaci
def fib(n):
    if n == 1 or n == 0:
        return 1
    else:
        return fib(n-1) + fib(n-2)

print(fib(5))
#Time complexity will be >2^n because below tree is not balanced
```

```python
#Problem no 11:Show 2 power table number upto given/input term
def power(num):
    if num < 1:
        return 0
    elif num == 1:
        print(1)
        return 1
    else:
        prev = power(num//2)
        curr = prev*2
        print(curr)
        return curr
#Time complexity is O(logn)
```

```python
#Problem no 13 : calculate modulus or remainder of a number when divided by a number
def mod(a, b):
    if b <= 0:
      return -1
    div = a // b
    return a-div*b

print(mod(5, 3))
#Time complexity O(n)
```

```python
#Problem no 14:it sums the digits
def sum_digits(num):
    sum = 0
    while (num > 0):
        sum += num % 10
        num //= 10
    return sum

sum_digits(123)
#time complexity is O(logn)
```
