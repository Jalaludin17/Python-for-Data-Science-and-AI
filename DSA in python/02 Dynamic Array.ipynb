```python
import array

my_array = array.array('i', [1, 2, 3, 4, 5])

# Start the formatted string with opening bracket
formatted_string = "["

# Loop through each item and add it to the string
for i in range(len(my_array)):
    formatted_string += str(my_array[i])

    # Add comma and space unless it's the last item
    if i < len(my_array) - 1:
        formatted_string += ", "

# Add closing bracket at the end
formatted_string += "]"

# Print the result
print(formatted_string)
```

* Python lists can store different data types (heterogeneous), since they just store references.

```python
#class
import ctypes

#constructor
class MeraList:
  def __init__(self):
    self.size=1
    self.n=0
    self.A=self.__make_array(self.size*2)

#len function
  def __len__(self):
    return self.n

#print function
  def __str__(self):
    result=""
    for i in range(self.n):
      result=result+str(self.A[i])+','
    return '['+ result[:-1] +']'

#index
  def __getitem__(self, index):
    if 0<=index<self.n:
      return self.A[index]
    return "IndexError:Index out of range"

#find
  def find(self, item):
    for i in range(self.n):
      if self.A[i]==item:
        return i
      return f"{item} not in the list"
#pop
  def pop(self):
    if self.n==0:
      return "list is empty"
    else:
      self.n=self.n-1
      print(self.A[self.n])

#clear
  def clear(self):
    self.n=0
    self.size=1
    print(self.A)

#insert
  def insert(self, pos, item):
    if self.size==self.n:
      self.__resize_array(self.size*2)
    for i in range(self.n, pos, -1):
      self.A[i]=self.A[i-1]
    self.A[pos]=item
    self.n+=1

#delete
  def __delitem__(self, pos):
    if 0 <= pos < self.n:
      for i in range(pos, self.n - 1):
        self.A[i] = self.A[i + 1]
      self.n -= 1
    else:
      raise IndexError("Index out of range")

#remove
  def remove(self, item):
    pos=self.find(item)
    if type(pos)==int:
      self.__delitem__(pos)
    self.n-=1
    return pos

#append
  def append(self, item):
    if self.n==self.size:
      self.__resize_array(self.size*2)
    self.A[self.n]=item
    self.n+=1

#resize
  def __resize_array(self, new_capacity):
    B=self.__make_array(new_capacity)
    self.size=new_capacity
    for i in range(self.n):
      B[i]=self.A[i]
    self.A=B

#make array
  def __make_array(self, capacity):
    return (capacity*ctypes.py_object)()
```

### Other functions

```python
class DynamicArray:
    def __init__(self):
        self.size = 1
        self.n = 0
        self.A = self.__make_array(self.size)

    def __len__(self):
        return self.n

    def __str__(self):
        return "[" + ", ".join(str(self.A[i]) for i in range(self.n)) + "]"


    def __getitem__(self, index):
        if isinstance(index, slice):
            # Slicing support
            start, stop, step = index.indices(self.n)
            sliced = DynamicArray()
            for i in range(start, stop, step):
                sliced.append(self.A[i])
            return sliced
        elif isinstance(index, int):
            # Negative indexing support
            if index < 0:
                index += self.n
            if 0 <= index < self.n:
                return self.A[index]
            raise IndexError("Index out of bounds")
        else:
            raise TypeError("Invalid argument type")

    def append(self, item):
        if self.n == self.size:
            self.__resize(2 * self.size)
        self.A[self.n] = item
        self.n += 1

    def extend(self, iterable):
        for item in iterable:
            self.append(item)

    def sort(self):
        # In-place sorting using Timsort (Python built-in)
        temp = [self.A[i] for i in range(self.n)]
        temp.sort()
        for i in range(self.n):
            self.A[i] = temp[i]

    def min(self):
        if self.n == 0:
            raise ValueError("Array is empty")
        minimum = self.A[0]
        for i in range(1, self.n):
            if self.A[i] < minimum:
                minimum = self.A[i]
        return minimum

    def max(self):
        if self.n == 0:
            raise ValueError("Array is empty")
        maximum = self.A[0]
        for i in range(1, self.n):
            if self.A[i] > maximum:
                maximum = self.A[i]
        return maximum

    def sum(self):
        total = 0
        for i in range(self.n):
            total += self.A[i]
        return total

    def merge(self, other):
        if not isinstance(other, DynamicArray):
            raise TypeError("Argument must be a DynamicArray")
        result = DynamicArray()
        result.extend(self)
        result.extend(other)
        return result

    def __resize(self, new_capacity):
        B = self.__make_array(new_capacity)
        for i in range(self.n):
            B[i] = self.A[i]
        self.A = B
        self.size = new_capacity

    def __make_array(self, capacity):
        return [None] * capacity
```
